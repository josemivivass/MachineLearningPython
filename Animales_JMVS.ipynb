```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from google.colab import drive

# Montar Google Drive
drive.mount('/content/drive')

# Definir ruta en Google Drive
directorio_datos = "/content/drive/MyDrive/Colab Notebooks/2EVA Antonio/animales"

# Definir parГЎmetros
tamano_lote = 32

# Generadores de datos con mГЎs augmentaciones
generador_datos_entrenamiento = ImageDataGenerator(
    rescale=1./255,
    horizontal_flip=True,
    rotation_range=40,
    zoom_range=0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    brightness_range=[0.8, 1.2],
    shear_range=0.2,
    fill_mode='nearest'
)

generador_datos_validacion = ImageDataGenerator(rescale=1./255)

entrenamiento = generador_datos_entrenamiento.flow_from_directory(
    os.path.join(directorio_datos, "entrenamiento"),
    batch_size=tamano_lote,
    class_mode='categorical'
)

validacion = generador_datos_validacion.flow_from_directory(
    os.path.join(directorio_datos, "validacion"),
    batch_size=tamano_lote,
    class_mode='categorical'
)

# ConstrucciГіn del modelo CNN sin tamaГ±o de imagen fijo
modelo = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(None, None, 3)),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.5),
    layers.Dense(128, activation='relu'),
    layers.Dense(4, activation='softmax')  # 4 clases de salida
])

# CompilaciГіn del modelo
modelo.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

epocas = 300
historial = modelo.fit(entrenamiento, validation_data=validacion, epochs=epocas, verbose=1)

# Guardar el modelo en Google Drive
ruta_guardado_modelo = os.path.join(directorio_datos, "modelo_300epocas.h5")
modelo.save(ruta_guardado_modelo)
print(f"Modelo guardado en: {ruta_guardado_modelo}")
```
