```python
#Importo librerias
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score , mean_absolute_error
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#Leer los datos excell
fich = "datos1.xls"
df = pd.read_excel(fich)
df
```

```python
#Numero de celdas
df.size
```

```python
#Estructura filas y columnas
df.shape
```

```python
#Filas
df.shape[0]
```

```python
#Columnas
df.shape[1]
```

```python
#Tipos de datos
df.info()
```

```python
#Columnas
df.columns
```

```python
#Pasar object a categoricas
df['SEXO'] = pd.Categorical(df['SEXO'])
df['FUMADOR'] = pd.Categorical(df['FUMADOR'])
df['POBLACION'] = pd.Categorical(df['POBLACION'])
df['NACIONALIDAD'] = pd.Categorical(df['NACIONALIDAD'])
df['ESTUDIOS'] = pd.Categorical(df['ESTUDIOS'])
```

```python
#Tipos de datos
df.info()
```

```python
#Columnas dummies
gender_dummy = pd.get_dummies(df['SEXO'], prefix='SEXO')
df = pd.concat([df, gender_dummy], axis=1)


df = df.drop(['SEXO'], axis = 1)

df.info()
```

```python
#Cat codes
df['FUMADOR'] = df['FUMADOR'].cat.codes

df.info()
```

```python
#LabelEncoder
encoder = LabelEncoder()
df["POBLACION"] = encoder.fit_transform(df["POBLACION"])
df
```

```python
#Cat codes
df['NACIONALIDAD'] = df['NACIONALIDAD'].cat.codes
df['ESTUDIOS'] = df['ESTUDIOS'].cat.codes
df.info()
```

```python
#Dividir el df en X e Y
X = df.drop('ENFERMEDAD', axis=1)
Y = df["ENFERMEDAD"]
```

```python
#Normalizar X
scaler = StandardScaler()
x1= scaler.fit_transform(X)
```

```python
#Aplicar train test split
X_train, X_test, y_train, y_test = train_test_split(x1, Y, test_size=0.2, random_state=0)
```

#Regresión Lineal

```python
from sklearn import linear_model
```

```python
# Creo un modelo de regresión lineal
modelo = linear_model.LinearRegression()
```

```python
X_train.shape
```

```python
# Entreno el modelo con los datos (X,Y)
modelo.fit(X_train.reshape(32,7), y_train)
```

```python
X_test.shape
```

```python
# Podemos predecir usando el modelo
y_pred = modelo.predict(X_test.reshape(8,7))
```

```python
# Por último, calculamos el error cuadrático medio y el estadístico R^2
print('Error cuadrático medio: %.2f' % mean_squared_error(y_test, y_pred))
print('Estadístico R_2: %.2f' % r2_score(y_test, y_pred))
```

```python
mean_absolute_error(y_test, y_pred)
```

#KNN vecinos

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
```

```python
clf = KNeighborsRegressor()
params = {"n_neighbors": range(1, 20)}
```

```python
clf_gs = GridSearchCV(clf, params, cv=3)
clf_gs.fit(X_train.reshape(32,7), y_train)
print('El mejor estimador es: {}'.format(clf_gs.best_estimator_))
```

```python

y_axis_pred = clf_gs.best_estimator_.predict(X_test.reshape(8,7))
```

```python
# Por último, calculamos el error cuadrático medio y el estadístico R^2
print('Error cuadrático medio: %.2f' % mean_squared_error(y_test, y_axis_pred))
print('Estadístico R_2: %.2f' % r2_score(y_test, y_axis_pred))
```

```python
mean_absolute_error(y_test, y_axis_pred)
```

#Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import cross_val_score, RepeatedKFold
```

```python
# Defino un modelo
model = RandomForestRegressor()
```

```python
import statistics
import numpy as np
```

```python
# Evalúo el modelo
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model, X_train.reshape(32,7), y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1, error_score='raise')
# Obtengo una métrica de rendimiento
print(u'Obtengo el error absoluto medio (MAE): %.3f, desviación %.3f' % (statistics.mean(n_scores), np.std(n_scores)))
```

```python
# Ajusto el modelo a los datos
model.fit(X_train.reshape(32,7),y_train)
y_prediccion = model.predict(X_test.reshape(8,7))
```

```python
# Por último, calculamos el error cuadrático medio y el estadístico R^2
print('Error cuadrático medio: %.2f' % mean_squared_error(y_test, y_prediccion))
print('Estadístico R_2: %.2f' % r2_score(y_test, y_prediccion))
```

```python

mean_absolute_error(y_test, y_prediccion)
```
